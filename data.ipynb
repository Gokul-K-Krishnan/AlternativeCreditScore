```
df = spark.read.option("inferSchema", "true").option("header", "true").parquet("dbfs:/mnt/loan/per/")
display(df)
```

```
df = df.withColumnRenamed("Seller.Name", "Seller_Name")
df = df.withColumnRenamed("Product.Type", "Product_Type")
```

```
from pyspark.sql.functions import col , column

df = df.withColumn("ORIG_RT", col("ORIG_RT").cast("double"))
df = df.withColumn("ORIG_AMT", col("ORIG_AMT").cast("double"))
df = df.withColumn("ORIG_TRM", col("ORIG_TRM").cast("double"))
df = df.withColumn("OLTV", col("OLTV").cast("double"))
df = df.withColumn("OCLTV", col("OCLTV").cast("double"))
df = df.withColumn("NUM_BO", col("NUM_BO").cast("double"))
df = df.withColumn("DTI", col("DTI").cast("double"))
df = df.withColumn("MI_PCT", col("MI_PCT").cast("double"))
df = df.withColumn("CSCORE_C", col("CSCORE_C").cast("double"))
df = df.withColumn("MI_TYPE", col("MI_TYPE").cast("double"))
df = df.withColumn("CSCORE_B", col("CSCORE_B").cast("double"))
```

```
from pyspark.sql.types import DateType
from pyspark.sql.functions import col, udf
from datetime import datetime

func =  udf (lambda x: datetime.strptime(x, '%m/%Y'), DateType())

df = df.withColumn('ORIG_DTE', func(col('ORIG_DTE')))
df = df.withColumn('FRST_DTE', func(col('FRST_DTE')))
```
